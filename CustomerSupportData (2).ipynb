```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visual settings for seaborn
sns.set(style='whitegrid')
```

```python
# Sample data for customer support
data = {
    "Ticket ID": ["001", "002", "003", "004", "005"],
    "Customer Name": ["Teodora V.", "Jane S.", "Chris P.", "Emilia D.", "Michael B."],
    "Issue Type": ["Billing", "Technical Support", "Account Issue", "Technical Support", "Billing"],
    "Issue Status": ["Closed", "Open", "In Progress", "Closed", "Closed"],
    "Response Time (Hours)": [2, 1, 3, 2, 4],
    "Resolution Time (Hours)": [5, None, None, 8, 6],
    "Customer Rating": [4, None, None, 5, 3]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the first few rows of the DataFrame
df.head()
```

```python
# Calculate average response and resolution times
avg_response_time = df["Response Time (Hours)"].mean()
avg_resolution_time = df["Resolution Time (Hours)"].mean()

print(f"Average Response Time: {avg_response_time} hours")
print(f"Average Resolution Time: {avg_resolution_time} hours")

# Count the number of issues by status
status_counts = df["Issue Status"].value_counts()
print("\nNumber of Issues by Status:")
print(status_counts)
```

```python
# Bar plot for the number of issues by status
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="Issue Status", palette="viridis")
plt.title("Number of Issues by Status")
plt.xlabel("Issue Status")
plt.ylabel("Count")
plt.show()
```

```python
# Bar plot for average response and resolution times
plt.figure(figsize=(8, 5))
times = pd.DataFrame({
    "Metric": ["Average Response Time", "Average Resolution Time"],
    "Hours": [avg_response_time, avg_resolution_time]
})
sns.barplot(data=times, x="Metric", y="Hours", palette="viridis")
plt.title("Average Times in Customer Support")
plt.ylabel("Hours")
plt.show()
```

```python
# Pie chart for the distribution of issue types
issue_type_counts = df["Issue Type"].value_counts()
issue_type_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, figsize=(6, 6), colormap='viridis')
plt.title("Distribution of Issue Types")
plt.ylabel("")
plt.show()
```
